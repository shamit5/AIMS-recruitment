```python
import os
import random
import shutil

# Paths
dataset_path = "/content/VOC2012_dark_YOLO_format"  # original dataset
images_path = os.path.join(dataset_path, "images")
labels_path = os.path.join(dataset_path, "labels")

output_path = "/content/voc_dark_split"
train_img_dir = os.path.join(output_path, "train", "images")
train_lbl_dir = os.path.join(output_path, "train", "labels")
val_img_dir = os.path.join(output_path, "val", "images")
val_lbl_dir = os.path.join(output_path, "val", "labels")

# Create folders
for d in [train_img_dir, train_lbl_dir, val_img_dir, val_lbl_dir]:
    os.makedirs(d, exist_ok=True)

# List all images
all_images = [f for f in os.listdir(images_path) if f.endswith((".jpg", ".png", ".jpeg"))]
random.shuffle(all_images)

# 80% train, 20% val split
split_idx = int(0.8 * len(all_images))
train_files = all_images[:split_idx]
val_files = all_images[split_idx:]

# Copy images + labels
def copy_files(file_list, target_img_dir, target_lbl_dir):
    for f in file_list:
        img_src = os.path.join(images_path, f)
        lbl_src = os.path.join(labels_path, f.replace(".jpg", ".txt").replace(".png", ".txt"))
        if os.path.exists(lbl_src):  # only if label exists
            shutil.copy(img_src, target_img_dir)
            shutil.copy(lbl_src, target_lbl_dir)

copy_files(train_files, train_img_dir, train_lbl_dir)
copy_files(val_files, val_img_dir, val_lbl_dir)

print("âœ… Split complete!")
print("Train images:", len(os.listdir(train_img_dir)))
print("Val images:", len(os.listdir(val_img_dir)))
```

```python

from ultralytics import YOLO
import clip
import torch
from PIL import Image
import numpy as np
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import os

device = "cuda" if torch.cuda.is_available() else "cpu"

# 1ï¸âƒ£ Load models
yolo = YOLO("/content/custom_path/best_yolo.pt")  # your fine-tuned YOLO
clip_model, clip_preprocess = clip.load("ViT-B/32", device=device, jit=False)
clip_model.eval()

# 2ï¸âƒ£ Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# 3ï¸âƒ£ Input text query
query_text = input("Enter query text: ")  # e.g., "a person"

# 4ï¸âƒ£ Run YOLO detection
results = yolo(image_path)

# 5ï¸âƒ£ Find best matching crop using CLIP
best_score = -1
best_crop = None
best_box = None
all_crops = []

for result in results:
    boxes = result.boxes.xyxy.cpu().numpy()
    img = result.orig_img
    for box in boxes:
        x1, y1, x2, y2 = map(int, box)
        cropped_img = img[y1:y2, x1:x2]

        # Skip empty crops
        if cropped_img.size == 0:
            continue

        pil_img = Image.fromarray(cropped_img)

        # CLIP preprocessing
        image_input = clip_preprocess(pil_img).unsqueeze(0).to(device)
        text_input = clip.tokenize([query_text]).to(device)

        with torch.no_grad():
            image_features = clip_model.encode_image(image_input)
            text_features = clip_model.encode_text(text_input)
            image_features /= image_features.norm(dim=-1, keepdim=True)
            text_features /= text_features.norm(dim=-1, keepdim=True)
            similarity = (image_features @ text_features.T).item()

        all_crops.append({
            "image": pil_img,
            "array": cropped_img,
            "score": similarity,
            "box": box
        })

        if similarity > best_score:
            best_score = similarity
            best_crop = pil_img
            best_box = box
            best_crop_array = cropped_img

# 6ï¸âƒ£ Output results
if not all_crops:
    print("No objects detected.")
else:
    # Create output directory
    os.makedirs("output", exist_ok=True)

    # A. Show best matching crop
    print(f"\nBest match (similarity: {best_score:.3f}):")
    plt.figure(figsize=(5,5))
    plt.imshow(best_crop)
    plt.axis('off')
    plt.show()

    # B. Save best crop to file
    best_crop_path = "output/best_crop.jpg"
    best_crop.save(best_crop_path)
    print(f"Saved best crop to: {best_crop_path}")

    # C. Save best crop as numpy array
    np.save("output/best_crop.npy", best_crop_array)
    print(f"Saved best crop array to: output/best_crop.npy")

    # D. Show all crops with scores (sorted)
    print("\nAll detected objects (sorted by similarity):")
    sorted_crops = sorted(all_crops, key=lambda x: x["score"], reverse=True)

    for i, crop_info in enumerate(sorted_crops[:5]):  # Show top 5
        plt.figure(figsize=(3,3))
        plt.imshow(crop_info["image"])
        plt.title(f"Rank {i+1}\nScore: {crop_info['score']:.3f}")
        plt.axis('off')
        plt.show()

        # Save each crop
        crop_path = f"output/crop_{i+1}.jpg"
        crop_info["image"].save(crop_path)

    # E. Show original image with best box
    img_with_box = cv2.imread(image_path)
    img_with_box = cv2.cvtColor(img_with_box, cv2.COLOR_BGR2RGB)
    x1, y1, x2, y2 = map(int, best_box)
    cv2.rectangle(img_with_box, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img_with_box, f"{best_score:.2f}", (x1, y1-10),
               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)

    plt.figure(figsize=(10,10))
    plt.imshow(img_with_box)
    plt.title(f"Original image with best matching object (score: {best_score:.3f})")
    plt.axis('off')
    plt.show()
```
